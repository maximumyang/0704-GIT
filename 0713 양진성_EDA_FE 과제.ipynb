- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
df1 = df.copy()
# df.shape[1]
```

```python
#Q12


missing_df2 = df1.isnull().sum() / len(df1)
# missing_df2

for i in range(df.shape[1]):
    if missing_df2[i] >= 0.25:
        df1 = df1.drop([df.columns[i]], axis=1)
    elif missing_df2[i] <0.1:
        if df.dtypes[i]!='object':
            df1[df.columns[i]].fillna(df1[df.columns[i]].median, inplace=True)
        else:
            df1[df.columns[i]].fillna(df1[df.columns[i]].mode()[0], inplace=True)

df1 = df1.dropna() #25%이상, 10%이상은 처리해주고 나머지에 대해서는 한꺼번에 dropna
        
        
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1)
#host_is_superhost 열에 대해 라벨 인코딩 진행
#슈퍼호스트이냐의 여부를 1,0으로 표현하게 함.
#2가 나오는 걸로 보아 결측치로 취급하여 드롭하든 대체하여야 할 것. 

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoded = encoder.fit_transform(df['host_is_superhost'])
df['host_is_superhost'] = encoded
df.host_is_superhost.value_counts()

```

```python
#Q13-2)
# 'review_scores_rating'를 보면 대부분 100점 몰려있음을 알 수 있음. -> 차별성이 없다는 것.
# 따라서 scaling 진행하여 비슷한 80~100 사이에서 차별성 있는 점수를 부여해보겠다는 것.
#얼추 정규분포를 따르므로 표준정규화할 것
sns.kdeplot(df['review_scores_rating'])

from sklearn.preprocessing import StandardScaler
sd_scaler = StandardScaler()

sd_scaler = sd_scaler.fit_transform(df['review_scores_rating'].values.reshape(-1,1))
df['review_scores_rating'] = sd_scaler

df['review_scores_rating']
```

```python
#Q13-3)
df2 = df.copy()
#응답률이 얼마나 빠르냐에 따라 '빠름', '느림' 표시
# sns.kdeplot(df['host_response_rate'].str[:-1].astype('float')) #대부분 100에 몰려있기는 함.
df2['host_response_rate'] = df2['host_response_rate'].str[:-1].astype('float') #우선 퍼센트로 표시되어 문자열인 것을 숫자로 변환
df2.dropna(subset=['host_response_rate'], inplace=True)
```

```python
response_pace = []  #느릴땐 0 빠를땐 1 
for i in df2['host_response_rate']:
    if i < 92:  #평균보다 느리면 0 부여, 빠르면 1 부여
        response_pace.append(0)
    else:
        response_pace.append(1)


df2['response_pace'] = response_pace
```
